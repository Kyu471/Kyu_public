영화 X 영화 (장르 유사도) 매트릭스

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movie_similarity = cosine_similarity(genres_matrix, genres_matrix)
```

```python
movie_similarity = pd.DataFrame(movie_similarity, index=genres_matrix.index, columns=genres_matrix.index)
movie_similarity
```

평점 예측

```python
def predict_ratings(similarity, ratings):
  #평점과 유사도의 곱을 계산
  ratings_dot_similarity = ratings.dot(similarity)

  #유사도의 합을 계산(0으로 나누느 것을 방지하기 위해 매우 작은 값을 더함)
  sum_similarity = ratings.apply(lambda x: x.replace(0, np.nan)).notna().dot(similarity).replace(0, np.nan) + 1e-10

  #평점 예측을 계산
  predicted_ratings = ratings_dot_similarity.div(sum_similarity)

  return predicted_ratings

predicted_ratings = predict_ratings(movie_similarity, KNN_user_item_df)

predicted_ratings
```
